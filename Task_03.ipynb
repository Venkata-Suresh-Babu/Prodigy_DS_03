```python
category = bankdata_df.select_dtypes(include='object').columns
print(category)

numbers = bankdata_df.select_dtypes(exclude='object').columns
print(numbers)
```

```python
# Calculate the number of rows and columns for subplots
num_plots = len(category)
numbers_rows = (num_plots + 1) // 2  # Add 1 and divide by 2 to round up for odd numbers
numbers_cols = 2

# Create a new figure
plt.figure(figsize=(10, 15))  # Adjust the figure size as needed

# Loop through each feature and create a countplot
for i, feature in enumerate(category, 1):
    plt.subplot(numbers_rows, numbers_cols, i)
    sns.countplot(x=feature, data=bankdata_df, color='#34a853')
    plt.title(f'Bar Plot of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)

# Adjust layout to prevent overlap of subplots
plt.tight_layout()
plt.show()
```
